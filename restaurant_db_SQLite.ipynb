# challenge 11

 change the favorite dish for a customer by joining two tables 

````python
# explore the favorite Dish table
query = 'SELECT *   From Dishes  '
df = pd.read_sql_query( query, connection)
df.head(2)
````

````python
# explore the favorite Dish for Cleo 
query = 'SELECT cus.* , di.Name  From Customers cus  Join Dishes di ON cus.FavoriteDish = di.DishID WHere FirstName = "Cleo" AND LastName = "Goldwater"'
df = pd.read_sql_query( query, connection)
df
````

````python
# change the favorite dish for Cleo  Goldwater to Quinoa Salmon SaLad instaad of Fiesta Family Platter
Cursor.executescript("""UPDATE Customers
SET FavoriteDish  = (Select  DishID from  Dishes where Name = "Quinoa Salmon Salad") 
WHERE FirstName = 'Cleo' AND LastName = "Goldwater"; """)
# customerID is unique for the customer 
query = 'SELECT * FROM  Customers  where FirstName = "Cleo"AND LastName = "Goldwater" '  
df = pd.read_sql_query( query, connection)
df.head(2)
````
